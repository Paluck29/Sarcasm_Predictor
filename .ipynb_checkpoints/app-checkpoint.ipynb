```python
@app.route('/')
def home():
     return render_template('index.html')
```

```python
@app.route('/resultsform', methods=['POST'])
def resultsform():
    
    sentence = [x for x in request.form.values()]
    sen = sentence[0]
    features = name.Testing(sentence)
    print(features)
    
    predicted_v = model1.predict(features)
    output1 = predicted_v[0]
    print(predicted_v)
    if(predicted_v[0]==1):
        output1="Sarcastic"
    elif(predicted_v[0]==0):
        output1="Not sarcastic"
    print(output1)
    
    predicted_v = model2.predict(features)
    output2 = predicted_v[0]
    print(predicted_v)
    if(predicted_v[0]==1):
        output2="Sarcastic"
    elif(predicted_v[0]==0):
        output2="Not sarcastic"
    print(output2)
    
    predicted_v = model3.predict(features)
    output3 = predicted_v[0]
    print(predicted_v)
    if(predicted_v[0]==1):
        output3="Sarcastic"
    elif(predicted_v[0]==0):
        output3="Not sarcastic"
    print(output3)
    
    predicted_v = model4.predict(features)
    output4 = predicted_v[0]
    print(predicted_v)
    if(predicted_v[0]==1):
        output4="Sarcastic"
    elif(predicted_v[0]==0):
        output4="Not sarcastic"
    print(output4)
         
    return render_template('resultsform.html', sentence=sen, predicted_text_l=output1, predicted_text_n=output2, predicted_text_r=output3, predicted_text_s=output4)
```

```python
if __name__ == "__main__":
    from waitress import serve
    serve(app, host="0.0.0.0", port=8080)
```
